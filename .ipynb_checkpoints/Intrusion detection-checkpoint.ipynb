```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from sklearn.metrics import log_loss,plot_confusion_matrix,plot_roc_curve,accuracy_score
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.tree import DecisionTreeClassifier
filterwarnings("ignore")
```

<h2 style="color:red;">Reading the Dataset</h2>

```python
#reading the features
#get this file from https://www.kaggle.com/mrwellsdavid/unsw-nb15
from IPython.display import display,HTML
features_data = pd.read_csv("NUSW-NB15_features.csv",encoding = "latin-1",index_col = "No.")
HTML(features_data.to_html())
```

```python
features = features_data.Name.values
```

```python
data = pd.read_csv("UNSW-NB15_1.csv",names=features)
```

```python
display(data.info())
display(data.head(3))
```

<h2 style="color:red;">Understanding Target variables</h2>

```python
attack_category_dist = data.attack_cat.value_counts()
label_dist = data.Label.value_counts()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.barplot(attack_category_dist.values,attack_category_dist.index)
plt.title("values in attack category")
plt.subplot(1,2,2)
sns.barplot(label_dist.index,label_dist.values)
plt.title("values in Label")
plt.show()
```

From the above graph we can see that most of the value have 0 Labels means not attack and also there are imbalanced data point for each attack

<h2 style="color:red;">Removing redundant and not useful feature<h2>

There are some feature which changes from each infrastructure like source IP address, source port number, destination IP address, and destination port number so we remove this features

There are two feature record start time and record last time which is redundant beacuse we already have record duration feature which is extracted from subtracting record start time from record last time

```python
data = data.drop(['srcip','sport','dstip','dsport','Stime','Ltime','Label'],axis = 1)
```

<h2 style="color:red;">Splitting Train and test</h2>

Instead of simple random splitting i will use stratify strategy for splitting the train and test. because we have imbalanced target data.

```python
X = data.drop("attack_cat",axis = 1)
y = data['attack_cat'].fillna("Not Attack")
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y)
```

```python
print(f"Shape of X train is {X_train.shape}")
print(f"Shape of X test is {X_test.shape}")
print(f"Shape of y train is {y_train.shape}")
print(f"Shape of y test is {y_test.shape}")
```

<h2 style="color:red;">Understand the category Colums and encode them</h2>

```python
object_col = data.select_dtypes(include="object").columns
```

```python
object_col
```

```python
print("No of unique value in proto column - ",len(data.proto.unique()))
print("No of unique value in state column - ",len(data.state.unique()))
print("No of unique value in service column - ",len(data.service.unique()))
```

For encoding categorical data we can use either one hot encoding. but one draw back of one hot encoding is it creates a n-1 dimensional vector of each categorical column where n is number of catergory in that column 

So instead of using one hot encoding i will encode using response coding :  I will use the probability estimate of target variable for 

```python
category_feature_train = pd.DataFrame()
category_feature_test = pd.DataFrame()
categorical_features = object_col.drop("attack_cat")
unique_attack = y.unique()
for features in categorical_features:
    for attack in unique_attack:   
        status =(X_train[y_train==attack][features].value_counts()/X_train[features].value_counts()).fillna(0)
        category_feature_train[features + "_"+attack] = X_train[features].map(lambda c : status[c])
        category_feature_test[features + "_"+attack] = X_test[features].map(lambda c : status[c] if c in status else 1/len(unique_attack))
```

```python
#reduce the dimension to visualize
sample_point = category_feature_test.assign(y = y_test).sample(10000)
X_ = TSNE().fit_transform(sample_point.drop("y",axis =1 ))
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(X_[:,0],X_[:,1],marker = "X",alpha = .5,hue = sample_point['y'],cmap = "dark")
plt.axis("off")
plt.title("Visualiation of Categorical encoded data in 2D space",fontsize= 24)
plt.show()
```

<h2 style="color:red;">Standarizing numerical features</h2>

I am using column standarization on numerical features except binary ones

```python
numerical_feature = data.columns.drop(object_col)
numerical_feature
```

```python
#selecting binary feature 
binary_feature  = []
for f in numerical_feature:
    if sorted(data[f].unique()) == [0,1]:
        binary_feature.append(f)
print(binary_feature)
```

```python
#i dont want to standarize binary feature
numerical_feature = numerical_feature.drop(binary_feature)
```

```python
scaler = StandardScaler()
numerical_feature_encoded_train = scaler.fit_transform(X_train[numerical_feature])
numerical_feature_encoded_test = scaler.transform(X_test[numerical_feature])
```

```python
binary_feature_train = X_train[binary_feature]
binary_feature_test = X_test[binary_feature]
```

<h2>Concat All features<h2>

```python
X_train = np.hstack((category_feature_train.values,numerical_feature_encoded_train,binary_feature_train.values))
X_test = np.hstack((category_feature_test.values,numerical_feature_encoded_test,binary_feature_test.values))
```

```python
print("Final shape of Features")
print("Train - ",X_train.shape)
print("Test - ",X_test.shape)
```

<h2 style="color:red;">Model Building</h2>

```python
#funtion which save the accuracies and logg loss of all the models and plot the Confusion matrix
models =list()
log_losses = list()
accuracies = list()
def track_record(model_name,model,X_test = X_test):
    loss = log_loss(y_test,model.predict_proba(X_test))
    accuracy = accuracy_score(y_test,model.predict(X_test))
    print("log loss is - ",loss,"And Accuracy score is - ",accuracy)
    models.append(model_name)
    log_losses.append(loss)
    accuracies.append(accuracy)
    fig, ax = plt.subplots(figsize=(15, 10))
    plot_confusion_matrix(model, X_test, y_test, ax=ax)
    plt.show()
```

<h3>1. Random Model</h3>

Since this is multiclass classification problem i am using log loss to measure our model.
Log loss can range between 0 to infinite so i am creating a random model which help me to get the idea of worst log loss

```python
model = DummyClassifier(strategy="uniform")
model.fit(X_train,y_train)
```

```python
track_record("random_model",model)
```

<h3>2. Logistic Regression </h3>

Logistic Regression is basically a binary classifier. But we can use one vs all strategy to sollve multiclass problem.

```python
model = GridSearchCV(
    SGDClassifier(loss = "log",class_weight="balanced"),
    param_grid = {"alpha": [0.01,0.1,1,10,100]},
    scoring='neg_log_loss',
    cv= 3,
    n_jobs = -1,
    )
model.fit(X_train,y_train)
```

```python
track_record("Logistic Regression",model)
```

<h3>3. Naive Bayes<h3>

```python
model = GaussianNB() 
model.fit(X_train,y_train)
```

```python
track_record("Naive Bayes",model)
```

<h3>4 . Decision Tree</h3>

```python
model = GridSearchCV(
    DecisionTreeClassifier(class_weight="balanced"),
    param_grid = {"max_depth": [4,10,20,None]},
    scoring='neg_log_loss',
    cv= 3,
    n_jobs = -1,
    )
model.fit(X_train,y_train)
```

```python
track_record("Decision Tree",model)
```

<h4>Getting Feature importance from DT</h4>

I am tring to get all non zero features importances index so we can use that in later

```python
model = model.best_estimator_
model.fit(X_train,y_train)
```

```python
important_feature_index = np.where(model.feature_importances_ != 0 )
```

<h3>5. Random Forest</h3>

```python
model = GridSearchCV(
    RandomForestClassifier(class_weight="balanced"),
    param_grid = {"n_estimators": [100,500,700]},
    scoring='neg_log_loss',
    cv= 3,
    n_jobs = 1,
    )
model.fit(X_train,y_train)
```

```python
track_record("Random Forest",model)
```

<h1 style="color:red;">Reduce Dimension LDA</h1>

I am using LDA to reduce the dimensions to n-1 where n is the number of classes

```python
lda = LinearDiscriminantAnalysis()
lda.fit(X_train,y_train)
X_train_ = lda.transform(X_train)
X_test_ = lda.transform(X_test)
X_train_.shape,X_test_.shape
```

```python
#logistic regression on reduced dataset
model = SGDClassifier(loss = "log",class_weight="balanced")
model.fit(X_train_,y_train)
```

```python
track_record("Log Regression after LDA",model,X_test=X_test_)
```

```python
#random forest in reduced dataset
model = RandomForestClassifier(n_estimators=500,class_weight="balanced",n_jobs = -1)
model.fit(X_train_,y_train)
```

```python
track_record("RF after LDA",model,X_test=X_test_)
```

```python
#knn in reduced dataset
model = KNeighborsClassifier(weights="distance")
model.fit(X_train_,y_train)
```

```python
track_record("KNN after LDA",model,X_test=X_test_)
```

<h2 style="color:red;" >Modelling Based on Non Zero features importance given by Decision Tree</h2>

```python
X_train_ = X_train[:,important_feature_index[0]]
X_test_ = X_test[:, important_feature_index[0]]
X_train_.shape,X_test_.shape
```

```python
#logistic regression
model = SGDClassifier(loss = "log",class_weight="balanced")
model.fit(X_train_,y_train)
```

```python
track_record("LR based on DT feature imp.",model,X_test=X_test_)
```

```python
#random forest classifier
model = RandomForestClassifier(n_estimators=500,class_weight="balanced",n_jobs = -1)
model.fit(X_train_,y_train)
```

```python
track_record("RF based on DT feature imp",model,X_test=X_test_)
```
