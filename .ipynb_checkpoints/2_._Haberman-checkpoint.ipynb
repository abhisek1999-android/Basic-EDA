<b>Relevant Information:</b>
The dataset contains cases from a study that was conducted between
1958 and 1970 at the University of Chicago's Billings Hospital on
the survival of patients who had undergone surgery for breast
cancer.

<b>Number of Instances:</b> 306

<b>Number of Attributes:</b> 4 (including the class attribute)

<b>Attribute Information:</b>
<ol type ="i">
    <li>
Age of patient at time of operation (numerical)
    </li>
    <li>
Patient's year of operation (year - 1900, numerical)
    </li>
    <li>
Number of positive axillary nodes detected (numerical)
    </li>
    <li>
    
Survival status (class attribute)<br>
1 = the patient survived 5 years or longer<br>
2 = the patient died within 5 year
    </li>
</ol>
Missing Attribute Values: None

```python
plt.figure(figsize=(14,5))
plt.suptitle("Violin Plots",size = 25)
plt.subplot(1,3,1)
sns.violinplot(x = "status",y = "age",data = data)
plt.ylabel("Age of patient at time of operation")
plt.subplot(1,3,2)
sns.violinplot(x = "status",y ="year",data = data)
plt.ylabel("Patient's year of operation")
plt.subplot(1,3,3)
sns.violinplot(x = "status",y ="nodes",data=data)
plt.ylabel("Number of positive axillary nodes detected")
plt.show()
```
